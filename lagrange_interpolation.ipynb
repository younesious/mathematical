# Lagrange Polynomial Interpolation


## Introduction

This Jupyter notebook demonstrates Lagrange Polynomial Interpolation. The user is prompted to input the number of data points and the corresponding x and y values. 
The Lagrange interpolation polynomial is then calculated and plotted along with the original data points.

### Lagrange Interpolation Function

The `lagrange_interpolation` function takes arrays `x` and `y` representing data points and a set of x-values `xp` at which to interpolate. It returns the corresponding y-values using Lagrange polynomial interpolation.

Importing the necessary packages

```julia
using Plots
```

define lagrange interpolation function here

```julia
function lagrange_interpolation(x, y, xp)
    n = length(x)
    yp = 0.0
    
    for i in 1:n    
        p = 1.0
        
        for j in 1:n
            if i != j
                p *= (xp - x[j]) / (x[i] - x[j])
            end
        end
        
        yp += p * y[i]
    end
    
    return yp
end
```

get user input

```julia
println("Enter number of data points: ")
n = parse(Int, readline())
```

Initialize x and y coordinates with 0 values

```julia
x = zeros(n)
y = zeros(n)
```

get points from user one by one

```julia
println("Enter data for x and y: ")
for i in 1:n
    print("x[$i] = ")
    x[i] = parse(Float64, readline())
    
    print("y[$i] = ")
    y[i] = parse(Float64, readline())
end
```

### Interpolation

The `range` function is used to generate x-values for interpolation, and the Lagrange interpolation function is applied to obtain y-values.

```julia
# Interpolation points
xp = range(minimum(x), stop=maximum(x), length=1000)

# Interpolated values
yp = [lagrange_interpolation(x, y, point) for point in xp]
```

### Plotting

The resulting Lagrange interpolation curve is plotted along with the original data points using the `Plots` package.

```julia
plot(xp, yp, label="F(x)")
scatter!(x, y, color=:red, label="Data Points")
title!("Lagrange Polynomial Interpolation")
xlabel!("x")
ylabel!("y")
```
